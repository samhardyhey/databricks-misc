## Installation Methods

There are two ways to install Spark NLP:
1. **Via Cluster Libraries (Recommended for production)**: Install at cluster level through UI
2. **Via Notebook (Quick setup for experimentation)**: Install using magic commands below


## Method 1: Install via Cluster Libraries (Recommended)

1. Go to your cluster settings
2. Click on **Libraries** tab
3. Click **Install New**
4. Select **PyPI** → Enter `spark-nlp==6.2.0` → Click **Install**
5. Click **Install New** again
6. Select **Maven** → Enter coordinates `com.johnsnowlabs.nlp:spark-nlp_2.12:6.2.0` → Click **Install**
7. Restart the cluster

This method installs libraries at the cluster level and they persist across notebook sessions.
